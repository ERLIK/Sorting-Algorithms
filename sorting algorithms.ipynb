```python
def mergeSort(dizi):
    if len(dizi) > 1:
        mid = len(dizi)//2
        left = dizi[:mid]
        right = dizi[mid:]
        mergeSort(left)
        mergeSort(right)
        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                dizi[k] = left[i]
                i += 1 
            else:
                dizi[k] = right[j]
                j += 1 
            k += 1

def mergeSortTimer():
    timer = list()
    for i in butunListeler:
        startPoint = dt.timestamp(dt.now())
        
        mergeSort(i)
        endPoint = dt.timestamp(dt.now())
        timer.append(endPoint - startPoint)
        
    return timer
```

```python
def quickSort(diziListe):
    uzunluk = len(diziListe)                    
    if len(diziListe) < 2:                      
        return diziListe                       
    
    anlik_pozisyon = 0                          

    for i in range(1, uzunluk):               
        if diziListe[i] <= diziListe[0]:     
            anlik_pozisyon += 1               
            diziListe[i], diziListe[anlik_pozisyon] = diziListe[anlik_pozisyon], diziListe[i]   

    diziListe[0], diziListe[anlik_pozisyon] = diziListe[anlik_pozisyon], diziListe[anlik_pozisyon] 

    left = quickSort(diziListe[0:anlik_pozisyon])          
    right = quickSort(diziListe[anlik_pozisyon+1:uzunluk]) 

    diziListe = left + [diziListe[anlik_pozisyon]] + right
    return diziListe

def quickSortTimer():
    timer = list()
    for i in butunListeler:
        startPoint = dt.timestamp(dt.now())
        
        quickSort(i)
        endPoint = dt.timestamp(dt.now())
        timer.append(endPoint - startPoint)
        
    return timer
```

```python
def heapify(dizi, n, i):
    largest = i
    left = 2*i+1
    right = 2*i+2
    
    if left < n and dizi[i] < dizi[left]:
        largest = left
    
    if right < n and dizi[largest] < dizi[right]:
        largest = right
    
    if largest != i:
        dizi[i], dizi[largest] = dizi[largest], dizi[i]
        heapify(dizi, n, largest) 
        
def heapSort(dizi):
    n = len(dizi)

    for i in range(n//2, -1 , -1):
        heapify(dizi, n, i)
    
    for i in range(n-1, 0, -1):
        dizi[i], dizi[0] = dizi[0], dizi[i]
        heapify(dizi, i, 0)
        
def heapifySortTimer():
    timer = list()
    for i in butunListeler:
        startPoint = dt.timestamp(dt.now())
        heapSort(i)
        endPoint = dt.timestamp(dt.now())
        timer.append(endPoint - startPoint)
        
    return timer
```

```python
def selectionSort(dizi):
    for i in range(len(dizi)-1):
        min_index = i
        for j in range(i+1, len(dizi)-1):
            if dizi[j] < dizi[min_index]:
                min_index = j
        dizi[i], dizi[min_index] = dizi[min_index], dizi[i]

def selectionSortTimer():
    timer = list()
    for i in butunListeler:
        startPoint = dt.timestamp(dt.now())
        selectionSort(i)
        endPoint = dt.timestamp(dt.now())
        timer.append(endPoint - startPoint)
        
    return timer
```

```python
def bubbleSort(dizi):
    for i in range(len(dizi)):
        for j in range(0, len(dizi) - i - 1):
            if dizi[j] > dizi[j + 1]:
                dizi[j],dizi[j+1] = dizi[j+1],dizi[j]

def bubbleSortTimer():
    timer = list()
    for i in butunListeler:
        startPoint = dt.timestamp(dt.now())
        bubbleSort(i)
        endPoint = dt.timestamp(dt.now())
        timer.append(endPoint - startPoint)
        
    return timer
```

```python
def insertionSort(dizi):
    for step in range(1, len(dizi)):
        key = dizi[step] 
        j = step - 1 
        while j >= 0 and key < dizi[j]: 
            dizi[j + 1] = dizi[j]   
            j = j - 1
        dizi[j + 1] = key

def insertionSortTimer():
    timer = list()
    for i in butunListeler:
        startPoint = dt.timestamp(dt.now())
        insertionSort(i)
        endPoint = dt.timestamp(dt.now())
        timer.append(endPoint - startPoint)
        
    return timer
```

```python
def shellSort(dizi):
    n = len(dizi)
    interval = n // 2
    while interval > 0:
        for i in range(interval, n):
            temp = dizi[i]
            j = i 
            while j >= interval and dizi[j - interval] > temp:
                dizi[j] = dizi[j - interval]
                j -= interval

            dizi[j] = temp
        interval //= 2
        
def shellSortTimer():
    timer = list()
    for i in butunListeler:
        startPoint = dt.timestamp(dt.now())
        shellSort(i)
        endPoint = dt.timestamp(dt.now())
        timer.append(endPoint - startPoint)
        
    return timer
```
